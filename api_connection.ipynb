```python
import requests

url = "https://xeno-canto.org/api/3/recordings"

all_recordings = []


for page in range(1, 11): 


    params = {
        "query": 'cnt:"Colombia"', 
        "key" : "1cb9951b67dbd5be783fae428acf24ac773356fc"
    }

    headers = {
        "Authorization": "1cb9951b67dbd5be783fae428acf24ac773356fc",
        "Content-Type": "application/json",
    }

    response = requests.get(url, params=params, headers=headers)
    data = response.json()


    all_recordings.extend(data["recordings"])

    print(f"Página {page} descargada")

    print("Total registros:", len(all_recordings))
```

```python
import requests
import os

def download_audio(url, save_path):
    response = requests.get(url)
    if response.status_code == 200:
        with open(save_path, "wb") as f:
            f.write(response.content)

os.makedirs("audios", exist_ok=True)

for rec in all_recordings:
    url = rec["file"]
    label = rec["gen"] + "_" + rec["sp"]
    filename = f"audios/{label}_{rec['id']}.mp3"

    download_audio(url, filename)
```
